```python
import os
import cv2
from deepface import DeepFace

# Path folder input dan output
input_folder = r'Data\wajah\raw\Fadh'  # Ganti dengan path folder Alice
output_folder = r'Data\wajah\crop\Fadh'  # Output akan disimpan di folder Alice dalam folder output

# Buat folder output jika belum ada
os.makedirs(output_folder, exist_ok=True)

# Iterasi setiap file dalam folder
for file_name in os.listdir(input_folder):
    file_path = os.path.join(input_folder, file_name)

    # Pastikan file adalah gambar
    if file_name.lower().endswith(('.jpg', '.jpeg', '.png')):
        try:
            # Deteksi wajah
            detections = DeepFace.analyze(img_path=file_path, detector_backend='retinaface', actions=['emotion'])

            # Baca gambar
            img = cv2.imread(file_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # Loop melalui wajah yang terdeteksi
            for idx, face in enumerate(detections):
                region = face['region']
                x, y, w, h = region['x'], region['y'], region['w'], region['h']
                x, y = max(0, x), max(0, y)  # Validasi koordinat
                cropped_img = img[y:y + h, x:x + w]  # Crop gambar

                # Simpan hasil crop
                cropped_img_bgr = cv2.cvtColor(cropped_img, cv2.COLOR_RGB2BGR)  # Konversi ke BGR
                output_file = os.path.join(output_folder, f"{os.path.splitext(file_name)[0]}_crop_{idx + 1}.jpg")
                cv2.imwrite(output_file, cropped_img_bgr)
                print(f"Cropped face saved to {output_file}")

        except Exception as e:
            print(f"Error processing {file_path}: {str(e)}")
```

```python
import os

# Path folder database
database_path = r'Data\wajah\raw'
output_path = r'Data\wajah\crop'

# Buat folder output jika belum ada
os.makedirs(output_path, exist_ok=True)

for root, dirs, files in os.walk(database_path):
    for folder in dirs:
        folder_path = os.path.join(root, folder)
        output_folder_path = os.path.join(output_path, folder)

        # Buat folder output untuk setiap subfolder
        os.makedirs(output_folder_path, exist_ok=True)

        for file_name in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file_name)

            # Pastikan file adalah gambar
            if file_name.lower().endswith(('.jpg', '.jpeg', '.png')):
                try:
                    # Deteksi wajah
                    detections = DeepFace.analyze(img_path=file_path, detector_backend='retinaface', actions=['emotion'])

                    # Baca gambar
                    img = cv2.imread(file_path)
                    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

                    # Loop melalui wajah yang terdeteksi
                    for idx, face in enumerate(detections):
                        region = face['region']
                        x, y, w, h = region['x'], region['y'], region['w'], region['h']
                        x, y = max(0, x), max(0, y)  # Validasi koordinat
                        cropped_img = img[y:y + h, x:x + w]  # Crop gambar

                        # Penamaan file output dengan hanya nama folder
                        output_file = os.path.join(
                            output_folder_path,
                            f"{folder}_{file_name}"
                        )
                        cropped_img_bgr = cv2.cvtColor(cropped_img, cv2.COLOR_RGB2BGR)  # Konversi ke BGR
                        cv2.imwrite(output_file, cropped_img_bgr)
                        print(f"Cropped face saved to {output_file}")

                except Exception as e:
                    print(f"Error processing {file_path}: {str(e)}")
```

```python
# Path folder hasil cropping
cropped_folder = r'Data\wajah\crop'
preprocessed_folder = r'Data\wajah\preprocessed'

target_size = (224, 224)

# Buat folder output jika belum ada
os.makedirs(preprocessed_folder, exist_ok=True)

# Proses setiap gambar di folder hasil cropping
for root, dirs, files in os.walk(cropped_folder):
    for folder in dirs:
        folder_path = os.path.join(root, folder)
        output_folder_path = os.path.join(preprocessed_folder, folder)

        # Buat folder output untuk setiap subfolder
        os.makedirs(output_folder_path, exist_ok=True)

        for file_name in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file_name)

            # Pastikan file adalah gambar
            if file_name.lower().endswith(('.jpg', '.jpeg', '.png')):
                try:
                    # Baca gambar
                    img = cv2.imread(file_path)
                    if img is None:
                        print(f"Skipping {file_path}, not a valid image.")
                        continue

                    # Resize gambar ke ukuran target
                    resized_img = cv2.resize(img, target_size, interpolation=cv2.INTER_AREA)

                    # Simpan gambar hasil resize
                    output_file = os.path.join(output_folder_path, file_name)
                    cv2.imwrite(output_file, resized_img)
                    print(f"Preprocessed image saved to {output_file}")

                except Exception as e:
                    print(f"Error processing {file_path}: {str(e)}")
```
